```python
sns.boxplot(dataset['SalePrice'])
```

```python
# sales price is output
```

```python
# Considering easch and every features and consider nan values = 1 and going to plot diagram and plot with depend feature sales price
```

```python
data = dataset.copy()

for feature in features_with_nan:
    
    # Let's make the variable that indicate 1 if the observation was missing or 0 otherwise
    data[feature]=np.where(data[feature].isnull(),1,0)
    
    # Lets calculate the mean salesPrice where the information is missing or present. # Assuming there is outlier so we are using median
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.show()
```

 Here with the relation between the missing values and the dependent variable is clearly visible. So we need to replace these nan values with something meaningful which we will do in Feature Engineering Section

```python
## Temporal Variables(Eg: Datetime Variables)
[feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
```

```python
year_feature = [feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
```

```python
### Here we will compare the difference between all years features with salesprice
data = dataset.copy()
for feature in year_feature:
    if feature!='YrSold':
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

### Observation
Within all the yearwise information, It is clearly visible that the New houses are more costlier than that of older one.Looks like a pareto or power law distribution

### numerical variables - 2types
1. Continuous variable and Discrete Variable


```python
[feature for feature in numerical_feature if len(dataset[feature].unique())<=25]
```

```python
discrete_feature=[feature for feature in numerical_feature if len(dataset[feature].unique())<=25]
print(len(discrete_feature))
```

```python
dataset[discrete_feature].head()
```

```python
## Find the relationship between Discrete and Sales Price
data = dataset.copy()
for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.xlabel('SalePrice')
    plt.title(feature)
    plt.show()
    
```

```python
## There is a relationship between Discrete feature and sale price
```

```python
## continuous variable
[feature for feature in numerical_feature if feature not in discrete_feature+year_feature]
```

```python
continuous_feature = [feature for feature in numerical_feature if feature not in discrete_feature+year_feature]
print(len(continuous_feature))
```

```python
# we can analyze the continuous values by creating histogram to understand the distribution

data = dataset.copy()
for feature in continuous_feature:
    data.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.xlabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python

data = dataset.copy()
for feature in continuous_feature:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.title(feature)
    plt.show()
    
```

# EDA part 2

```python
#log normal distribution to normal distribution usin logarithmic transformation

data = dataset.copy()
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
                                
```

```python
## outlier
data = dataset.copy()
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
